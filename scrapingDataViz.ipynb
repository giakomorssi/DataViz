### Stacked bar chart for `categories` form tweet text corpus

```python
# Stacked bar chart for categories form tweet text corpus:

import re
import pandas as pd
import matplotlib.pyplot as plt

# Define the regular expression pattern for categories
pattern = r'py\w+' #change # with your hashtag

# Define a function to extract the categories from the tweet text
def extract_categories(text):
    return re.findall(pattern, text)

# Define a sample dataset
data = df['tweet_text']

# Extract the categories for each tweet and count their frequency
data = pd.DataFrame(data)
data['categories'] = data['tweet_text'].apply(extract_categories)
category_counts = data['categories'].apply(pd.Series).stack().value_counts()

# Create a stacked bar chart
fig, ax = plt.subplots(figsize=(20, 10))
category_counts.plot(kind='bar', stacked=True, ax=ax)

# Set the chart title and axis labels
ax.set_title('Category Distribution in Tweet Text')
ax.set_xlabel('Categories')
ax.set_ylabel('Frequency')

# Show the chart
plt.show()
```

### Tweet Length Distribution in Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
data = df['tweet_text']

# Get lengths of each item in the dataset
lengths = [len(item) for item in data]

# Create density plot using Seaborn
sns.kdeplot(lengths, color='#8856a7', fill=True, alpha=.3, linewidth=2)

# Customize plot
plt.xlabel('Length', fontsize=14, fontweight='bold', color='#404040')
plt.ylabel('Density', fontsize=14, fontweight='bold', color='#404040')
plt.title('Tweet Length Distribution in Dataset', fontsize=16, fontweight='bold', color='#404040')

# Add gridlines
plt.grid(axis='y', color='#d4d4d4', linestyle='--', linewidth=1, alpha=0.5)

# Remove top and right spines
sns.despine(top=True, right=True)

# Add a text box with dataset information
info_text = f"Dataset: {len(data)} items"
plt.text(0.02, 0.98, info_text, transform=plt.gca().transAxes,
         fontsize=12, fontweight='bold', color='#404040', ha='left', va='top', bbox=dict(facecolor='white', alpha=0.8))

# Add a legend
plt.legend(['Lengths'], loc='best', fontsize=12, frameon=True, edgecolor='#404040', facecolor='white')

# Set figure background color
plt.gca().set_facecolor('#f5f5f5')

# Show plot
plt.show()

```

### Histogram of tweet lengths

```python
# Histogram of tweet lengths
import matplotlib.pyplot as plt
import seaborn as sns

# sample data
tweet_lengths = df['tweet_text'].str.len()

sns.set_style("darkgrid")

# create figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# create histogram in the background
mean_length = sum(tweet_lengths) / len(tweet_lengths)
ax.axvline(mean_length, color='red', linestyle='--', label=f'Mean length: {mean_length:.2f}')

# create main histogram
sns.histplot(tweet_lengths, bins=35, kde=True, ax=ax)

# set labels and title
ax.set_xlabel('Tweet length')
ax.set_ylabel('Frequency')
ax.set_title('Histogram of tweet lengths')

# add legend
ax.legend()

# show plot
plt.show()
```

### Total User Followers and User Friends by Query

```python
import plotly.graph_objects as go

df_grouped = df.groupby("query")[["user_followers", "user_friends"]].sum().reset_index()

fig = go.Figure(data=[
    go.Bar(name='User Followers', x=df_grouped["query"], y=df_grouped["user_followers"],
           hovertext=df_grouped["user_followers"], hoverinfo="text+y"),
    go.Bar(name='User Friends', x=df_grouped["query"], y=df_grouped["user_friends"],
           hovertext=df_grouped["user_friends"], hoverinfo="text+y")
])
fig.update_layout(barmode='stack', title="Total User Followers and User Friends by Query",
                  xaxis=dict(title="Query", title_font=dict(size=16), tickfont=dict(size=14)),
                  yaxis=dict(title="Total Count", title_font=dict(size=16), tickfont=dict(size=14)),
                  font=dict(size=14))
fig.show()
```
